# Operadores de decisición

```python
x = int(input("Escribe un número"))
```

```python
if x == 5:
    print("Has escrito el número 5")
    
```

```python
x = int(input("Escribe un número: "))
if x < 5:
    print("Has escrito un número menor que 5")
elif x < 10:
    print("Has escrito un número menor que 10 y mayor o igual a 5")
else:
    print("El número que has escrito es mayor o igual que 10")
```

```python
x = int(input("Escribe un número"))
if x>0 and x < 10:
    print("Has escrito un número en el rago de [0,10]")
```

```python
x = int(input("Escribe un número: "))
if x < 0 or x > 10:
    print("Tú numero escrito no se encuentra dentro del rango [0,10]")
```
