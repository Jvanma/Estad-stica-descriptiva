### Funciones trigonométricas

```python
math.sin(180)   # Calcular el seno "Ojo, debe especificarse en radianes"
```

```python
math.cos(math.pi) # Calcular el coseno de Pi
```

```python
math.tan(math.pi/2) # Tangente de pi/2
```

```python
math.asin(1) # Arcoseno, ¿cual sería el ángulo de un seno que vale 1? El resultado es en RADIANES
```

```python
math.acos(1) # Arcocoseno, ¿cual sería el ángulo de un coseno cuyo valor vale 1? El resultado es en RADIANES
```

```python
math.atan(1) # Arcotangente, ¿cual es el ángulo cuyo valor de tangente vale 1?   El resultado es en RADIANES
```

```python
math.degrees(math.atan(1)) # Esta función convierte de radianes a grados
```

```python
math.radians(60) # Convierte grados a radianes
```

```python
math.cos(math.radians(60)) # esta función calcula el coseno de 60 grados.
```

```python
math.hypot(3, 4) # Norma euclídea para conocer la longitud (largo) de un vector por el ancho y el alto 
```

```python
math.sqrt(3**2+4**2)      # Implementación de un hypot
```

```python
math.atan2(4, 3) # El ángulo que forma el vector con el eje horizontal. 
# En la función primero se indica el eje de las equis y segundo argumento el eje y.
# El resultado siempre es un ángulo entre -pi y pi
```
